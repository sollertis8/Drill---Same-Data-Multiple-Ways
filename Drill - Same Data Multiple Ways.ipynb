```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import kendalltau
%matplotlib inline

# Loading our Dataset into a Dataframe
df = pd.read_csv('World-Religion-Dataset-National-Religion-Dataset.csv', thousands=',')

# Slicing portions of the data we want to work with and assigning them to variables
christian = df['CHRSGEN'][0:14]
pop = df['POP'][0:14]
year = df['YEAR'][0:14]
# w = np.array([[christian], [year]])
dfchrs = pd.DataFrame()
dfchrs['Year'] = year
dfchrs['Christian'] = christian
dfchrs['Population'] = pop
dfchrs['Percentage of Christians'] = christian / pop * 100

crp_norm = (christian / pop) * 100


sns.set()

ax = sns.barplot(year, crp_norm)
ax.set(ylabel = 'Percentage of Christians')

sns.lmplot(x="Year", y="Percentage of Christians",
           data=dfchrs, scatter_kws={"s": 50, "alpha": 1})


sns.lmplot(x="Year", y="Christian",
           data=dfchrs, scatter_kws={"s": 50, "alpha": 1})


plt.tight_layout(h_pad=3)
```

The bar chart is useful as it gives us all of the information we need about christians in the U.S. over the years.  The  first line chart could be useful as it shows us the distributions of christians over time but because it doesn't start at zero, the graph can be misinterpreted as the differences appear more dramatic. The last chart isn't very useful as it is'nt clear what the numbers on the y-axis represent.

```python
sns.set()

chrs_dist = sns.distplot(christian, color="m")
chrs_dist.set(xlabel="Christian")
```

This graph could be more useful if we converted the y-axis to percentages.

```python
sns.set()

chrs = sns.distplot(christian, hist=False, color="g", kde_kws={"shade": True})
chrs.set (xlabel='Christian')
```
